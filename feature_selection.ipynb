```python
selected_indicators = indicators_no_dup + indicators_aid_effectiveness_no_dup + indicators_urban_developmen_no_dup + indicators_trade_no_dup + indicators_financial_sector_no_dup
selected_indicators = list(dict.fromkeys(i for i in selected_indicators))
selected_indicators
```

```python
len(selected_indicators)
```

```python
filtered_countries = result_dataframe[result_dataframe['Country Name'].isin(eu + g20)]

filtered_countries_indicators = filtered_countries[filtered_countries.index.isin(selected_indicators, level=2)]

filtered_countries_indicators.head()
```

```python
plt.figure(figsize=(15,15))
filtered_countries_indicators.iloc[:, 31:].isnull().sum(axis=0).plot.bar()
```

```python
filtered_countries_indicators_years = filtered_countries_indicators.iloc[:, 31:]
filtered_countries_indicators_years['Country Name'] = filtered_countries_indicators.iloc[:, 0]
filtered_countries_indicators_years.head()
```

```python
for category_name, category_data in filtered_countries_indicators_years.groupby(level=1):
    print(f'{category_name}:')
    for indicator_name, _ in category_data.groupby(level=2):
        print(f'\t{indicator_name}')
```

```python
filtered_countries_indicators_years.to_csv('data/selected_data.csv')
```
