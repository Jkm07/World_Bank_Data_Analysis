Dla wybranych wskaźników liczba Nanów dla każdego kraju

```python
import pandas as pd

countries_low_NaN_count = result_dataframe[result_dataframe.index.isin(selected_indicators, level=2)].iloc[:, 32:-1].isnull().sum(axis = 1).groupby('Country Code').sum().sort_values()

countries_names = result_dataframe['Country Name'].groupby('Country Code').first()

countries_low_NaN_count = pd.concat([countries_names, countries_low_NaN_count], axis = 1)

countries_low_NaN_count = countries_low_NaN_count.rename(columns={0: 'NaN Count'})

countries_low_NaN_count
```

Wszystkie kraje w kolejności liczby NaNów

```python
for country in countries_low_NaN_count.sort_values('NaN Count')[:110].iterrows():
    print(f"{country[1]['Country Name']}: {country[1]['NaN Count']}")
```

Kraje uzupełniające i liczba NaNów

```python
complementary_countries = []
for country in countries_low_NaN_count[~countries_low_NaN_count['Country Name'].isin(eu + g20)].sort_values('NaN Count')[:59].iterrows():
    print(f"{country[1]['Country Name']}: {country[1]['NaN Count']}")
    complementary_countries.append(country[1]['Country Name'])
```
