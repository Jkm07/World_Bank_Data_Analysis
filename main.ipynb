```python
result_dataframe = da.concat_all_datasets(result)
result_dataframe
```

Czy indeks jest unikalny

```python
result_dataframe.index.is_unique
```

Nieunikalne wskaźniki (nazwa tego samego wskaźnika w róźnych kategoriach)

```python
result_dataframe.groupby(['Country Code', 'Indicator Name']).size().groupby(['Indicator Name']).max().loc[lambda count : count > 1]
```

Przykład nieunikalnego wskaźnika

```python
result_dataframe[result_dataframe.index.get_level_values('Indicator Name') == 'Access to electricity (% of population)'].groupby('Category Name').size()
```

Liczba Nan dla każdego wskaźnika i kraju

```python
result_dataframe.iloc[:, 31:].isnull().sum(axis=1).sort_values(ascending=False)
```

Liczba wskaźników, które mają wypełnione wszystkie wartości

```python
result_dataframe.iloc[:, 31:].isnull().sum(axis=1).loc[lambda x : x == 0].sort_values(ascending=False)
```

Liczba wskaźników, które nie mają żadnej wartości

```python
result_dataframe.iloc[:, 31:].isnull().sum(axis=1).loc[lambda x : x == 33].sort_values(ascending=False)
```

Histogram ilości Nan wartości dla wszystkich indikatorów

```python
plt.figure(figsize=(15,15))
sns.histplot(data=result_dataframe.iloc[:, 31:].isnull().sum(axis=1), discrete = True)
```

Wykres Nan wartości dla każdego roku

```python
plt.figure(figsize=(15,15))
result_dataframe.iloc[:, 31:].isnull().sum(axis=0).plot.bar()
```
