```python
import pandas as pd
```

```python
def get_data_set(file_name: str, indicator_name: str):
    data = pd.read_csv(file_name, header = 2)
    data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
    data = data[(data['Indicator Name'] == indicator_name)]
    data = data.drop(columns=['Indicator Name', 'Indicator Code'])
    data = data.set_index('Country Code')
    return data
```

```python
get_data_set("data/trade.csv", 'Travel services (% of commercial service exports)')
```
